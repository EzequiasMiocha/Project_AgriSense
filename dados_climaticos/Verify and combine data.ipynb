```python
#Importando bibliotecas
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
```

```python
import os

os.listdir()
```

```python
#Acessando as bibliotecas
ds1 = xr.open_dataset('data_stream-moda_stepType-avgad.nc', engine='netcdf4')
ds2 = xr.open_dataset('data_stream-moda_stepType-avgua.nc', engine='netcdf4')

ds = xr.merge(
    [ds1, ds2],
    join="outer",      # permite diferenças nas dimensões
    compat="override"
)
```

```python
print(ds)
```

```python
# Transformando para um CSV

df = ds.to_dataframe().reset_index()

#Salvado os dados para um novas analises

df.to_csv('Dados_clima_2018_2013.csv', index= False)
```
