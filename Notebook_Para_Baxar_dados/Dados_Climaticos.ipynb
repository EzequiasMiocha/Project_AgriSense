```python
import cdsapi
import os

dataset = "reanalysis-era5-single-levels-monthly-means"

request = {
    "product_type": ["monthly_averaged_reanalysis"],
    "variable": [
        "10m_u_component_of_wind",
        "10m_v_component_of_wind",
        "2m_temperature",
        "total_precipitation",
        "surface_solar_radiation_downwards",
        "volumetric_soil_water_layer_1",
        "volumetric_soil_water_layer_2"
    ],
    "year": [
        "2018", "2019", "2020",
        "2021", "2022", "2023"
    ],
    "month": [
        "01", "02", "03",
        "04", "05", "06",
        "07", "08", "09",
        "10", "11", "12"
    ],
    "time": ["00:00"],
    "data_format": "netcdf",
    "download_format": "unarchived",
    "area": [-21.41, 31.39, -24.76, 33.79]
}

client = cdsapi.Client()

print("Iniciando download...")

try:
    client.retrieve(dataset, request).download()
    
    print("Download concluído com sucesso!")

    # Notificação no Ubuntu (opcional)
    os.system('notify-send "Download CDSAPI" "Download concluído com sucesso!"')

except Exception as e:
    print("Erro durante o download:", e)
    os.system('notify-send "Download CDSAPI" "Erro durante o download!"')
```
