```python
import cdsapi
from pathlib import Path

# Pasta de saída
output_folder = Path("dados_climaticos")
output_folder.mkdir(exist_ok=True)

output_file = output_folder / "era5_mensal_2018_2023.nc"

# Tempo
years = [str(year) for year in range(2018, 2024)]
months = [f"{month:02d}" for month in range(1, 13)]

# Variáveis corrigidas
variables = [
    "2m_temperature",
    "total_precipitation",
    "10m_u_component_of_wind",
    "10m_v_component_of_wind",
    "surface_solar_radiation_downwards",
    "volumetric_soil_water_layer_1",
    "volumetric_soil_water_layer_2"
]

# Área de Gaza
area = [-24.5, 33.4, -24.8, 33.6]

dataset = "reanalysis-era5-single-levels-monthly-means"

request = {
    "product_type": "monthly_averaged_reanalysis",
    "variable": variables,
    "year": years,
    "month": months,
    "time": "00:00",
    "format": "netcdf",
    "area": area
}

client = cdsapi.Client()

print("Baixando dados...")

client.retrieve(dataset, request, str(output_file))

print(f"Dados guardados em: {output_file}")
```
